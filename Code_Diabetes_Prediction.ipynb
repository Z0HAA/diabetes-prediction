# Diabetes Prediction System

## Meta Description
A machine learning model that analyzes patient medical data to predict whether a **female patient** has diabetes, using all 8 diagnostic features from the Pima Indians Diabetes Database with MLP Neural Network and StandardScaler preprocessing.

## Introduction

**Definition:** A binary classification ML task that predicts whether a female patient aged 21 or older is Diabetic (1) or Not Diabetic (0) based on 8 medical input features.

**Application:** Used in clinical decision support, early diabetes screening, and healthcare data analytics specifically for female patients of Pima Indian heritage aged 21 and above.

**Dataset Note:** This model is trained on the Pima Indians Diabetes Database, originally from the National Institute of Diabetes and Digestive and Kidney Diseases. All patients in this dataset are **females at least 21 years old** of Pima Indian heritage. Therefore, predictions are only valid for this demographic.

## Step 1: Import Libraries

```python
"""
Purpose of this code:
----------------------
This script imports the required libraries for building and evaluating
a machine learning model for diabetes prediction. It sets up tools for:
- Data manipulation and loading
- Splitting data into training and testing sets
- Scaling features using StandardScaler
- Training an MLP Neural Network
- Evaluating prediction accuracy
- Displaying results in tabular form
"""

# Import numerical computing library
import numpy as np

# Import data manipulation library
import pandas as pd

# Import pickle for saving and loading trained models
import pickle

# Import train_test_split to divide dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Import StandardScaler to normalize feature values (required for MLP)
from sklearn.preprocessing import StandardScaler

# Import MLP Neural Network classifier
from sklearn.neural_network import MLPClassifier

# Import accuracy_score to evaluate model performance
from sklearn.metrics import accuracy_score

# Import PrettyTable for clean tabular output
from prettytable import PrettyTable

print("All libraries imported successfully!")
```

## Step 2: Load Sample Data

```python
"""
Purpose of this section:
-------------------------
Load the 100-instance sample dataset (50 Diabetic, 50 Not Diabetic) from
a CSV file into a pandas DataFrame and display it for review.
The dataset contains all 8 input features and 1 output column (Outcome).
"""

# Read the CSV file into a pandas DataFrame
sample_data = pd.read_csv("sample-data.csv")

# Configure pandas to display all rows and columns without truncation
pd.set_option("display.max_rows", None, "display.max_columns", None)

print("\n\nSample Data:")
print("============\n")
print(sample_data)
```

## Step 3: Understand and Pre-process Sample Data
### Step 3.1: Understand Sample Data

```python
"""
Purpose of this section:
-------------------------
Understand the dataset by printing:
1. All attribute (column) names
2. Total number of instances
3. Class distribution (Diabetic vs Not Diabetic)
4. Data types and basic statistics
"""

print("\n\nAttributes in Sample Data:")
print("==========================\n")
print(sample_data.columns.tolist())

print("\n\nNumber of Instances in Sample Data:", sample_data["Pregnancies"].count())
print("=============================================\n")

print("\nOutcome Distribution (0 = Not Diabetic, 1 = Diabetic):")
print("=========================================================\n")
print(sample_data["Outcome"].value_counts())

print("\n\nData Types of Each Column:")
print("===========================\n")
print(sample_data.dtypes)

print("\n\nBasic Descriptive Statistics:")
print("==============================\n")
print(sample_data.describe())
```

### Step 3.2: Pre-process Sample Data — StandardScaler

**Why StandardScaler is needed:**
- All 8 features have very different numeric ranges:
  - Insulin ranges from 0 to 846
  - DiabetesPedigreeFunction ranges from 0.08 to 2.42
  - Pregnancies ranges from 0 to 17
- MLP Neural Networks are sensitive to feature scale — features with large ranges dominate and cause poor predictions.
- StandardScaler transforms each feature to have **mean = 0** and **standard deviation = 1**, putting all features on equal footing.

Scaling is applied in **Step 6.3** after the train/test split (to prevent data leakage).

## Step 4: Feature Extraction
- All **8 features** from the dataset are used as inputs — no feature extraction needed.


## Step 5: Label Encoding
- All 8 input features are already **numeric** — no Label Encoding required for inputs.
- The output column (Outcome) is already **0 or 1** — no Label Encoding required for output.


## Step 6: Execute the Training Phase
### Step 6.1: Splitting Sample Data into Training Data and Testing Data

```python
"""
Purpose of this section:
-------------------------
Split the 100-instance dataset into:
- Training Data: 80 instances (80%) — used to TEACH the model
- Testing Data:  20 instances (20%) — used to CHECK model performance

random_state=0 : ensures reproducibility (same split every run)
shuffle=False  : keeps original row order before splitting
"""

# Load directly from sample data (no encoding needed)

# 80% training, 20% testing
training_data, testing_data = train_test_split(
    sample_data,
    test_size=0.2,
    random_state=0,
    shuffle=True
)

# Save both splits to CSV
training_data.to_csv("training-data.csv", index=False, header=True)
testing_data.to_csv("testing-data.csv",   index=False, header=True)

print("\n\nTraining Data (80 instances):")
print("==============================\n")
print(training_data)

print("\n\nTesting Data (20 instances):")
print("=============================\n")
print(testing_data)
```

### Step 6.2: Splitting Input Vectors and Outputs / Labels of Training Data

```python
"""
Purpose of this section:
-------------------------
Separate the training data into:
- Input Vectors (X_train): All 8 feature columns
- Output Labels (y_train): Outcome column (1=Diabetic, 0=Not Diabetic)
"""

print("\n\nInput Vectors (Feature Vectors) of Training Data:")
print("==================================================\n")
input_vector_train = training_data.iloc[:, :-1]
print(input_vector_train)

print("\n\nOutputs / Labels of Training Data:")
print("====================================\n")
print("  Outcome")
output_label_train = training_data.iloc[:, -1]
print(output_label_train)
```

### Step 6.3: Scale the Features using StandardScaler

```python
"""
Purpose of this section:
-------------------------
Apply StandardScaler to normalize all 8 input features.

Why scale AFTER splitting?
- We fit the scaler ONLY on training data (fit_transform).
- We then apply the same scaler to test data (transform only).
- This prevents data leakage — the model never sees test data statistics
  during training, which would give artificially inflated accuracy.

fit_transform(X_train) : learns mean/std from training data AND scales it
transform(X_test)      : applies the SAME learned scaling to test data
"""

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit on training data and scale it
input_vector_train_scaled = scaler.fit_transform(input_vector_train)

# Scale testing data using the SAME scaler (do NOT fit again)
input_vector_test  = testing_data.iloc[:, :-1]
input_vector_test_scaled = scaler.transform(input_vector_test)

print("\n\nScaled Input Vectors of Training Data:")
print("=======================================\n")
print(input_vector_train_scaled)

print("\n\nScaled Input Vectors of Testing Data:")
print("======================================\n")
print(input_vector_test_scaled)
```

### Step 6.4: Train the MLP Neural Network Classifier

```python
"""
Purpose of this section:
-------------------------
Train an MLP Neural Network on the scaled training data.

MLP learns patterns through interconnected layers of neurons. It outperforms SVC on this dataset
because it can capture non-linear relationships between medical features.

Parameters used:
- max_iter=1000   : allows enough training iterations to converge
- alpha=1         : L2 regularization to prevent overfitting on small dataset
- random_state=0  : ensures reproducible results
"""

print("\n\nTraining the MLP Neural Network Classifier on Training Data")
print("=============================================================\n")

# Initialize MLP Neural Network
mlp_model = MLPClassifier(max_iter=1000, alpha=1, random_state=0)

# Train the model on SCALED training data
mlp_model.fit(input_vector_train_scaled, np.ravel(output_label_train))

print("Model trained successfully!")
print("\nModel Parameters:")
print("=================\n")
print(mlp_model)
```

### Step 6.5: Save the Trained Model and Scaler

```python
"""
Purpose of this section:
-------------------------
Save both the trained MLP model AND the fitted StandardScaler to disk.

IMPORTANT: Both must be saved together because:
- The scaler learned the mean/std from training data.
- Any new input MUST be scaled using this SAME scaler before prediction.
- If we lose the scaler, the model cannot make correct predictions.
"""

# Save the trained MLP model
pickle.dump(mlp_model, open("mlp_trained_model.pkl", "wb"))

# Save the fitted scaler (must use same scaler for new predictions)
pickle.dump(scaler, open("scaler.pkl", "wb"))

print("\nMLP model saved as: mlp_trained_model.pkl")
print("Scaler saved as:    scaler.pkl")
```

## Step 7: Execute the Testing Phase
### Step 7.1: Splitting Input Vectors and Outputs/Labels of Testing Data

```python
"""
Purpose of this section:
-------------------------
Separate the testing data into input features and output labels.
(input_vector_test_scaled was already prepared in Step 6.3)
"""

print("\n\nInput Vectors (Feature Vectors) of Testing Data - Scaled:")
print("==========================================================\n")
print(input_vector_test_scaled)

print("\n\nOutputs / Labels of Testing Data:")
print("==================================\n")
print("  Outcome")
output_label_test = testing_data.iloc[:, -1]
print(output_label_test)
```

### Step 7.2: Load the Saved Model and Scaler

```python
"""
Purpose of this section:
-------------------------
Load the saved MLP model and StandardScaler from disk.
Both are needed to make predictions on new data.
"""

model  = pickle.load(open("mlp_trained_model.pkl", "rb"))
scaler = pickle.load(open("scaler.pkl", "rb"))
print("\nMLP model loaded successfully!")
print("Scaler loaded successfully!")
```

### Step 7.3: Evaluate the Machine Learning Model
#### Step 7.3.1: Make Predictions with the Trained Model on Testing Data

```python
"""
Purpose of this section:
-------------------------
Use the trained MLP model to make predictions on the scaled testing data.
Store predictions alongside actual values for comparison.
"""

# Make predictions on scaled test data
model_predictions = model.predict(input_vector_test_scaled)

# Suppress chained assignment warning
pd.options.mode.chained_assignment = None

# Add predictions as a new column to testing dataset
testing_data["Predictions"] = model_predictions

# Save predictions to CSV
testing_data.to_csv("model-predictions.csv", index=False, header=True)

print("\n\nPredictions Returned by mlp_trained_model:")
print("===========================================\n")
print(testing_data)
```

### Step 7.4: Calculate the Accuracy Score

```python
"""
Purpose of this section:
-------------------------
Calculate the accuracy of the MLP model on the testing dataset.
Accuracy = (Correct Predictions / Total Predictions) x 100
"""

model_accuracy_score = accuracy_score(
    testing_data["Outcome"],
    testing_data["Predictions"]
)

print("\n\nAccuracy Score:")
print("===============\n")
print(round(model_accuracy_score, 2))
print(f"\nThe MLP model correctly predicted {round(model_accuracy_score * 100, 1)}% of the 20 test instances.")
```

## Step 8: Execute the Application Phase
### Step 8.1: Take Input from User

```python
"""
Purpose of this section:
-------------------------
Ask the user to enter their 8 medical measurements for diabetes prediction.
"""

pregnancies_input = float(input("\nPlease enter number of Pregnancies (e.g., 0, 1, 2, 6): ").strip())
glucose_input     = float(input("\nPlease enter Glucose level in mg/dL (e.g., 85, 120, 148): ").strip())
bp_input          = float(input("\nPlease enter Blood Pressure in mm Hg (e.g., 60, 72, 80): ").strip())
skin_input        = float(input("\nPlease enter Skin Thickness in mm (e.g., 0, 20, 35): ").strip())
insulin_input     = float(input("\nPlease enter Insulin level in mu U/ml (e.g., 0, 80, 200): ").strip())
bmi_input         = float(input("\nPlease enter BMI value (e.g., 22.5, 33.6, 43.1): ").strip())
dpf_input         = float(input("\nPlease enter Diabetes Pedigree Function (e.g., 0.167, 0.627, 1.2): ").strip())
age_input         = float(input("\nPlease enter Age in years (e.g., 21, 35, 50): ").strip())
```

### Step 8.2: Convert User Input into Feature Vector

```python
"""
Purpose of this section:
-------------------------
Convert the user's 8 inputs into a structured DataFrame (feature vector)
that exactly matches the column format used during training.
"""

user_input = pd.DataFrame({
    "Pregnancies":              [pregnancies_input],
    "Glucose":                  [glucose_input],
    "BloodPressure":            [bp_input],
    "SkinThickness":            [skin_input],
    "Insulin":                  [insulin_input],
    "BMI":                      [bmi_input],
    "DiabetesPedigreeFunction": [dpf_input],
    "Age":                      [age_input]
})

print("\n\nUser Input Feature Vector:")
print("==========================\n")
print(user_input)
```

### Step 8.3: Scale the Feature Vector (Exactly Same as Scaling Applied to Sample Data)

```python
"""
Purpose of this section:
-------------------------
Scale the user's input using the SAME StandardScaler that was fitted
on the training data and saved in Step 6.5.

Using the saved scaler ensures the transformation is identical to
what was applied during training — critical for correct predictions.
"""

# Load the saved scaler
scaler = pickle.load(open("scaler.pkl", "rb"))

# Scale user input using the same scaler
user_input_scaled = scaler.transform(user_input)

print("\n\nUser Input Feature Vector (Original):")
print("======================================\n")
print(user_input)

print("\n\nUser Input Feature Vector (Scaled):")
print("=====================================\n")
print(user_input_scaled)
```

### Step 8.4: Load the Saved Model

```python
"""
Purpose of this section:
-------------------------
Load the saved MLP model from disk for making predictions on new user data.
"""

model = pickle.load(open("mlp_trained_model.pkl", "rb"))
print("\nMLP model loaded successfully!")
```

### Step 8.5: Model Prediction
#### Step 8.5.1: Apply Model on the Scaled Feature Vector and Return Prediction to the User

```python
"""
Purpose of this section:
-------------------------
Apply the trained MLP model to the scaled user input and display
whether the patient is predicted to be Diabetic or Not Diabetic.
"""

# Make prediction on scaled user input
predicted_outcome = model.predict(user_input_scaled)

# Translate numeric result to readable label
if predicted_outcome[0] == 1:
    prediction = "DIABETIC"
else:
    prediction = "NOT DIABETIC"

# Display result in PrettyTable
pretty_table = PrettyTable()
pretty_table.add_column("       ** Prediction **       ", [prediction])
print(pretty_table)
```
